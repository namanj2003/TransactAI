```python
data = []

for category, keywords in categories.items():
    for i in range(120):
        keyword = random.choice(keywords)

        if category == 'Income':
            variations = [
                keyword,
                f"{keyword} RECEIVED",
                f"{keyword}-{random.randint(10000, 99999)}",
                f"CR-{keyword}",
                f"{keyword} REF{random.randint(100, 999)}",
                f"INCOMING {keyword}",
                f"{keyword} FROM SENDER",
                f"CREDITED BY {keyword}"
            ]
        elif category == 'Funds Transfer':
            variations = [
                keyword,
                f"{keyword} TO ACCOUNT",
                f"{keyword}-{random.randint(10000, 99999)}",
                f"DR-{keyword}",
                f"{keyword} REF{random.randint(100, 999)}",
                f"OUTGOING {keyword}",
                f"{keyword} TO BENEFICIARY",
                f"DEBITED VIA {keyword}"
            ]
        elif category == 'Others':
             variations = [
                keyword,
                f"{keyword} - {random.randint(100, 999)}",
                f"MISC {keyword}",
                f"UNCATEGORIZED {keyword}",
                f"{keyword} TXN {random.randint(1000, 9999)}"
             ]
        else:
            variations = [
                keyword,
                f"{keyword} PAYMENT",
                f"TO {keyword}",
                f"{keyword} ONLINE",
                f"UPI-{keyword}",
                f"{keyword}-{random.randint(1000, 9999)}",
                f"{keyword} TXN",
                f"{keyword} REF{random.randint(100, 999)}"
            ]

        description = random.choice(variations)
        data.append({'description': description, 'category': category})

df = pd.DataFrame(data)

df = df.sample(frac=1, random_state=42).reset_index(drop=True)

df.to_csv('training_data_comphrensive.csv', index=False)

print(f"Created training data: {len(df)} samples")
print(f"Categories: {df['category'].nunique()}")
print(f"\nCategory distribution:")
print(df['category'].value_counts().sort_index())
print(f"\nSample data:")
print(df.head(20))
```
